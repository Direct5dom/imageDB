导入所需要的库

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn import datasets, linear_model, model_selection
```

添加一些打印命令来显示数据矩阵的Shape及其第一个元素：

```python
X, y = datasets.load_diabetes(return_X_y=True)
print(X.shape)
print(X[0])
```

作为响应返回的是一个元组。你正在做的是将元组的前两个值分别分配给 `X` 和 `y`。

你可以看到这个数据有 442 个项目，组成了 10 个元素的数组。

接下来，通过使用 numpy 的 `newaxis` 函数，选择上面数据集的一部分（第三列），将其排列到一个新数组中。我们将使用线性回归根据它确定的模式在此数据中的值之间生成一条线。

```python
X = X[:, np.newaxis, 2]
```

现在你已准备好绘制数据，你可以查看计算机是否可以帮助确定此数据集中数字之间的逻辑分割。为此你需要将数据 (X) 和目标 (y) 拆分为测试集和训练集。Scikit-learn 有一个简单的方法来做到这一点；你可以在给定点拆分测试数据。

```python
X_train, X_test, y_train, y_test = model_selection.train_test_split(X, y, test_size=0.33)
```

现在已准备好训练模型。加载线性回归模型并使用 `model.fit()` 使用 X 和 y 训练集对其进行训练：

```python
model = linear_model.LinearRegression()
model.fit(X_train, y_train)
```

然后，使用函数 `predict()`，使用测试数据创建预测。这将用于绘制数据组之间的线：

```python
y_pred = model.predict(X_test)
```

现在是时候在图中显示数据了。可以利用 Matplotlib 来完成此任务。

创建所有 X 和 y 测试数据的散点图，并使用预测在模型的数据分组之间最合适的位置画一条线。

```python
plt.scatter(X_test, y_test,  color='black')
plt.plot(X_test, y_pred, color='blue', linewidth=3)
plt.show()
```

至此，我们成功用Python构建了第一个线性回归模型，使用它创建了预测，并以图形模式显示出来。
